## Make data frame for fitting

We want to get data frames for indicators of a given period, returning all other indicators for this period, and lags for the given indicator

```python
def get_your_data(data , org_unit_ID):
    data_to_fit = data[data['org_unit_ID'] == org_unit_ID]
    data_to_fit['indicID'] = data_to_fit['data_element_name']  + data_to_fit['category_name']

    data_to_fit_c = data_to_fit['indicID'].value_counts()
    to_keep = data_to_fit_c[data_to_fit_c > 50] 
    data_to_fit = data_to_fit[data_to_fit['indicID'].isin(to_keep.index)]
    return data_to_fit
```

```python
from sklearn.preprocessing import Imputer

def get_and_transpose_data(data , org_unit_ID):
    extr_data = get_your_data(data , org_unit_ID)
    if len(extr_data) > 0:
        data_pivot =  extr_data.pivot_table(index = ['period'], columns = ['indicID'], 
                                            values = 'value' , aggfunc = sum)
   #     data_pivot = data_pivot.dropna()
#        print data_pivot.columns
#        imp = Imputer(missing_values='NaN', strategy='mean', axis=0 , copy = False)
#        imp.fit(data_pivot)
#        data_pivot = DataFrame(imp.fit_transform(data_pivot))
#        print data_pivot.head()
        data_pivot = data_pivot.fillna(0.0)
    if len(extr_data) == 0 :
        data_pivot = 'N'
    print('extracted data is' , len(data_pivot))
    return data_pivot
```
